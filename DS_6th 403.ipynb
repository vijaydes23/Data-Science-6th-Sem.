###  STEP 5: Data Visualization
-  Bar Chart  Bedrooms vs Average Rent

```python
import matplotlib.pyplot as plt

bedroom_rent = df.groupby('bedrooms')['rent_price'].mean()

plt.bar(bedroom_rent.index, bedroom_rent.values)
plt.xlabel("Bedrooms")
plt.ylabel("Average Rent")
plt.title("Bedrooms vs Average Rent")
plt.show()
```

-  Scatter Plot  Area vs Rent

```python
plt.scatter(df['area_sqft'], df['rent_price'])
plt.xlabel("Area (sqft)")
plt.ylabel("Rent Price")
plt.title("Area vs Rent Price")
plt.show()
```

- Correlation Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

# correlation matrix
corr = df.corr()

plt.figure(figsize=(10, 7))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of House Rent Dataset")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df.drop('rent_price', axis=1)
y = df['rent_price']

model = RandomForestRegressor()
model.fit(X, y)

importance = pd.Series(model.feature_importances_, index=X.columns)
importance.sort_values().plot(kind='barh')
plt.title("Feature Importance")
plt.show()
```
