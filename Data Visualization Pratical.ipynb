## ðŸ”¹ PRACTICAL 4
- Identify the purpose of data visualization for a given dataset and use a donut chart.
- Purpose of Data Visualization using Donut Chart

```python
import matplotlib.pyplot as plt

# Dataset
departments = ['BCA', 'BBA', 'MBA', 'MCA']
students = [120, 90, 70, 60]

# Donut Chart
plt.figure()
plt.pie(students, labels=departments, autopct='%1.1f%%')
centre_circle = plt.Circle((0,0), 0.6, fc='white')
plt.gca().add_artist(centre_circle)

plt.title("Department-wise Student Distribution (Donut Chart)")
plt.show()
```

## ðŸ”¹ PRACTICAL 11
- Draw a simple node-link diagram to show relationships (e.g:, friends in a class).
- Node-Link Diagram (friends in a class)

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create Graph
G = nx.Graph()

# Nodes
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

# Links (Friendship)
G.add_edges_from([
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'D'),
    ('C', 'E'),
    ('D', 'E')
])

# Draw Graph
plt.figure()
nx.draw(G, with_labels=True)
plt.title("Node-Link Diagram (Friends in Class)")
plt.show()
```

## ðŸ”¹ PRACTICAL 12
- Visualize the correlation between different numeric variables in a dataset using a color-coded matrix (heatmap).
- Heatmap (Correlation Matrix)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = {
    'Maths': [65, 70, 75, 80, 85],
    'Science': [60, 68, 74, 82, 88],
    'English': [70, 72, 78, 85, 90]
}

df = pd.DataFrame(data)
df
```

### Correlation

```python
corr = df.corr()
corr
```

### Heatmap

```python
plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

##ðŸ”¹ PRACTICAL 14
- Perform text visualization by counting word frequency fron a paragraph and plotting it.
- Text visualization (word frequency)

```python
import matplotlib.pyplot as plt
from collections import Counter

# Paragraph
text = """
Data visualization helps in understanding data easily.
Visualization makes complex data simple.
Data analysis and visualization are important in data science.
"""

# Word count
words = text.lower().split()
word_count = Counter(words)

# Plot
plt.figure()
plt.bar(word_count.keys(), word_count.values())
plt.xticks(rotation=45)
plt.title("Word Frequency Visualization")
plt.show()
```

## ðŸ”¹ PRACTICAL 15
- Create a simple multivariate chart using more than two variables (e.g., marks in three subjects).
- Multivariate chart (marks in three subjects)

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
students = ['A', 'B', 'C', 'D']
maths = [70, 80, 75, 90]
science = [65, 85, 78, 88]
english = [72, 82, 80, 92]

x = np.arange(len(students))

# Bar width
width = 0.25

plt.figure()
plt.bar(x - width, maths, width, label='Maths')
plt.bar(x, science, width, label='Science')
plt.bar(x + width, english, width, label='English')

plt.xticks(x, students)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Multivariate Chart (Three Subjects)")
plt.legend()
plt.show()
```

## ðŸ”¹ PRACTICAL 17
- Create a live-like data visualization using continuously updated values (manual update is allowed).
- Live-like data visualization (manual update allowed)

```python
import matplotlib.pyplot as plt
import time
import random

x = []
y = []

plt.figure()

for i in range(10):
    x.append(i)
    y.append(random.randint(10, 100))

    plt.cla()
    plt.plot(x, y, marker='o')
    plt.title("Live-like Data Visualization")
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.pause(0.5)

plt.show()
```
