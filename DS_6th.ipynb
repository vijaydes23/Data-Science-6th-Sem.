## **Pratical - 1**
- Create a simple bar chart using student marks or sales data and explain what it shows.

```python
batsman = pd.read_csv('/content/sharma-kohli.csv')

plt.bar(batsman['index'], batsman['V Kohli'])

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("V Kohli Runs per Year")

plt.show()
```

## **Pratical - 2**
- Draw a line chart to represent time-based data (monthly attendance or sales).

```python
batsman = pd.read_csv('/content/sharma-kohli.csv')

# labels & tittle
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',marker='o',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',marker='o',markersize=10)

plt.title('Rohit Vs Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Run Score')
```

## **Pratical - 3**
- Perform data representation vs data presentation using at least two chart types

### Chart1: Data Representation (Bar Chart)

```python
import matplotlib.pyplot as plt

# Bar chart for RG Sharma
plt.bar(batsman['index'], batsman['RG Sharma'])

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("RG Sharma Runs")

plt.show()
```

### Chart2: Data Presentation (Pie Chart)

```python
plt.pie(batsman['RG Sharma'], labels=batsman['index'], autopct='%1.1f%%')

plt.title("RG Sharma Runs Distribution")

plt.show()
```

## **Pratical - 4**
- Identify the purpose of data visualization for a dataset and use a donut chart.

```python
plt.pie(batsman['V Kohli'],labels=batsman['index'],autopct='%1.1f%%')

plt.title("V Kohli Runs Distribution")

# for Donut Chart
plt.gca().add_artist(plt.Circle((0,0), 0.4, color='white'))

plt.show()
```
